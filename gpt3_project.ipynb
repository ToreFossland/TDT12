```python
def GPT_Prompt(texts):
    ## Call the API key under your account (in a secure way)
    openai.api_key = api_key
    response = openai.Completion.create(
    engine="text-davinci-002",
    prompt =  texts,
    temperature = 0.8,
    top_p = 1,
    max_tokens = 512,
    frequency_penalty = 0.5,
    presence_penalty = 0.5
    )
    return response.choices[0].text
```

```python
def describe_lyrics(lyrics):
    query = """{} 
How would you describe the meaning of the lyrics above? Describe some of the symbolism. 
""".format(lyrics)
    return GPT_Prompt(query)
```

```python
def describe_artist(artist):
    query = """Write a short bio for {}. What is the essence of their music, their genre, and how would you describe their personality?""".format(artist)
    return GPT_Prompt(query).strip()

#about_artist = describe_artist("Travis Scott")
```

```python
def mood_lyrics(lyrics):
    query = """{}

What is the mood of the above lyrics?

The mood is""".format(lyrics)
    return GPT_Prompt(query)

#mood = mood_lyrics(lyrics1)
#print(mood)
```

```python
def artist_printer(artist, title, lyrics):
    print("Artist: ", artist)
    print("Title: ", title)
    print("Lyrics: ", lyrics) 
    print("\n")
    print("About: ", describe_artist(artist))
    print("\n")
    print("Mood: ", mood_lyrics(lyrics))

artist_printer(artists[1], song_titles[1], song_lyrics[1])
```

```python
def make_prompt(artist, description_artist, description_lyrics, mood):
    prompt = "artist: {}\nabout: {}\ndescription: {}\nmood {}\n\n###\n\n".format(artist, description_artist, description_lyrics, mood)
    return prompt

def make_completion(song_name, lyrics):
    completion = "[{}]\n\n{}...<!>".format(song_name, lyrics)
    return completion
```

```python
def to_json_format(artist, song_name, lyrics):
    description_artist = describe_artist(artist)
    description_lyrics = describe_lyrics(lyrics)
    mood = mood_lyrics(lyrics)
    prompt = make_prompt(artist, description_artist, description_lyrics, mood)
    completion = make_completion(song_name, lyrics)
    dictionary = {"prompt": prompt, "completion": completion}
    return dictionary
```

```python
def populate_json(artists, song_titles, song_lyrics):
    to_json = []
    for i in range(len(artists)):
        to_json.append(to_json_format(artists[i], song_titles[i], song_lyrics[i]))
    print(to_json)
    with open("./song_data.json", "w") as f:
        for item in to_json:
            f.write(json.dumps(item)+ "\n")
```

```python
populate_json(artists, song_titles, song_lyrics)
```
